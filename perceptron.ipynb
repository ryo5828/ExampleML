# 単純パーセプトロン
#### y = 0 (重み{x1} × 入力{x1} + 重み{w2} × 入力{x2} <= 閾値)
#### y = 1 (重み{x1} × 入力{x1} + 重み{w2} × 入力{x2} > 閾値)

```python
# ANDゲート
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7 # 重み, 閾値を格納
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    else:
        return 1
```

```python
# 出力
print('出力:', AND(0, 0))
print('出力:', AND(1, 0))
print('出力:', AND(0, 1))
print('出力:', AND(1, 1))
```

### バイアスの導入
#### y = 0 (バイアス{b} + 重み{x1} × 入力{x1} + 重み{w2} × 入力{x2} <= 閾値)
#### y = 1 (バイアス{b} + 重み{x1} × 入力{x1} + 重み{w2} × 入力{x2} > 閾値)

```python
# ライブラリのインポート
import numpy as np

def AND(x1, x2):
    x = np.array([x1, x2]) # 入力
    w = np.array([0.5, 0.5]) # 重み
    b = -0.7 # バイアス
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
# 出力
print('出力:', AND(0, 0))
print('出力:', AND(1, 0))
print('出力:', AND(0, 1))
print('出力:', AND(1, 1))
```

```python
# Not AND ゲート
def NAND(x1, x2):
    x = np.array([x1, x2]) # 入力
    w = np.array([0.5, -0.5]) # 重み
    b = 0.7 # バイアス
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
# ORゲート
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5 ])
    b = -0.2 # バイアス
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
# XORゲート
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
# 出力
print('出力:', XOR(0, 0))
print('出力:', XOR(1, 0))
print('出力:', XOR(0, 1))
print('出力:', XOR(1, 1))
```
